```python
import requests
import time  # Pour ajouter des pauses
import json  # Pour sauvegarder les données JSON

# URL de l'API pour chercher des offres
url_offres = 'https://api.emploi-store.fr/partenaire/offresdemploi/v2/offres/search'

# En-têtes avec ton token d'accès
headers = {
    'Authorization': f'Bearer {token}'  # Remplace par ton token
}

categories = ['Informatique', 'Data', 'Marketing', 'Vente', 'Ressources Humaines', 'Finance', 'Juridique', 'Ingenierie', 'Conseil', 'Education', 'Sante', 'Hotellerie', 'Media', 'Immobilier', 'E-commerce', 'Environnement', 'Agriculture', 'Transport']
nom_categorie = ['Informatique', 'Data', 'Marketing', 'Vente', 'Ressources_Humaines', 'Finance', 'Juridique', 'Ingenierie', 'Conseil', 'Education', 'Sante', 'Hotellerie', 'Media', 'Immobilier', 'E-commerce', 'Environnement', 'Agriculture', 'Transport']
# Paramètres de base pour la recherche

for i in range(len(categories)):
    category = categories[i]
    file_name = nom_categorie[i]

    print(f"--- Récupération des offres pour la catégorie : {category} ---")
    params = {
        'motsCles': category,  # Mots-clés à chercher
    }

    # Variables pour la pagination
    all_offres = []
    offset = 0
    limit = 150  # Taille maximale par requête selon la documentation
    max_results = 10000  # Limite de réponse
    keep = True

    while keep:
        # Définir la plage actuelle pour la pagination
        params['range'] = f'{offset}-{offset + limit - 1}'

        # Faire la requête GET
        response = requests.get(url_offres, headers=headers, params=params)
        print(f"Status code: {response.status_code}, Range: {params['range']}")

        # Pause entre les requêtes
        time.sleep(0.15)

        if response.status_code in [200, 206]:  # Requête réussie
            try:
                data = response.json()  # Convertir en JSON
                offres = data.get('resultats', [])  # Extraire la liste des offres
                print(len(offres))

                if offres:
                    all_offres.extend(offres)  # Ajouter les résultats pertinents
                    offset += limit
                    keep = offset < max_results and len(offres) > 0
                else:
                    print("Aucune offre supplémentaire trouvée.")
                    break
            except ValueError:
                print(f"Erreur de décodage JSON pour range {params['range']}")
                break
        elif response.status_code == 204:  # Pas de contenu
            print("Fin des résultats atteinte.")
            break
        else:
            print(f"Erreur : Status code {response.status_code}")
            break

    output_file = f'raw_json_file\offres_emploi_{file_name}.json'
    print(output_file)
    try:
        with open(output_file, 'w', encoding='utf-8') as f:
            json.dump(all_offres, f, ensure_ascii=False, indent=4)
        print(f"Les offres ont été sauvegardées dans le fichier '{output_file}'")
    except Exception as e:
        print(f"Erreur lors de la sauvegarde du fichier JSON : {e}")
```
